```python
from datasets import load_from_disk
from pytoda.smiles.smiles_language import SMILESTokenizer

def inspect_dataset(dataset_path, vocab_path, num_samples=1):
    # 1. 데이터셋 로드
    print(f"📂 Loading dataset from: {dataset_path}")
    ds = load_from_disk(dataset_path)
    
    # 2. 토크나이저 로드 (디코딩 확인용)
    tokenizer = SMILESTokenizer(vocab_file=vocab_path)
    
    print(f"📊 Total samples: {len(ds)}")
    print(f"📝 Columns: {ds.column_names}")
    print("-" * 50)

    # 3. 샘플 출력
    for i in range(min(num_samples, len(ds))):
        sample = ds[i]
        input_ids = sample['input_ids']
        labels = sample['labels']

        # 디코딩: 숫자 -> SMILES 문자열
        # pytoda의 경우 indexes_to_smiles 또는 직접 id_to_token 사용
        decoded_tokens = [tokenizer.index_to_token[idx] for idx in input_ids]
        decoded_smiles = "".join(decoded_tokens).replace("<PAD>", "").replace("<CLS>", "[CLS]").replace("<EOS>", "[EOS]")

        print(f"--- Sample {i} ---")
        print(f"📍 Input IDs (first 15): {input_ids[:10]}...")
        print(f"📍 Decoded SMILES: {decoded_smiles}")
        print(f"📍 Labels: {labels}")
        print(f"📍 Sequence Length: {len(input_ids)}")
        print()

if __name__ == "__main__":
    TASK = ['clintox','bbbp','bace','tox21','sider','muv','hiv','esol','lipo','freesolv']
    for task in TASK:
        DATASET_PATH = f"FT_Data/{task}_dataset"
        VOCAB_PATH = "../smiles_tokenizer_pytoda/vocab.json"

        inspect_dataset(DATASET_PATH, VOCAB_PATH)
```
