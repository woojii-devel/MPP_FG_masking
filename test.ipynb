```python
from datasets import load_from_disk
from pytoda.smiles.smiles_language import SMILESTokenizer

VOCAB_PATH = "smiles_tokenizer_pytoda/vocab.json"

tok = SMILESTokenizer(vocab_file=VOCAB_PATH, canonical=True, sanitize=True)
ds = load_from_disk("arrow_shards/shard_00000")

sample = ds[1]
smi = sample["smiles"]
token_atom = sample["token_atom"]

tokens = tok.smiles_tokenizer(smi)
tokens = ["<s>"] + tokens + ["</s>"]

for t, a in zip(tokens, token_atom):
    print(f"{t:>8} -> {a}")
```
